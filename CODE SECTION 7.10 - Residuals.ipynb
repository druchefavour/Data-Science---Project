#### Residual Plots
For linear relatonships, the ideal plot of residuals with each of the predictor should be a random scatter because we assume that the residuals are uncorrelated with the predictor variables. Any noticeable pattern in such plots indicates violation of linear relationship assumption.

Below are plots of the residuals with each of the predictor variables in the given data set. As we can see the plots are random scatter plots. The plots are indicating linearity.

Therefore, linear regression analysis will be applied to the data.

```python
%matplotlib inline
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
from scipy import stats
import matplotlib as mpl
import matplotlib.pyplot as plt
```
