Biogenic is created by methanogenic lifeforms that exist in shallow waters or landfills while further down in the earth thermogenic gas is created from organic material that has been buried for a great deal of time.

Natural gas is a major source of power generation, especially for heating and cooling systems in the U.S. Generally, the product is used to either power gas turbines or burned to create steam and power steam turbines instead. The fuel is often preferred to coal or oil since it is far cleaner and produces less greenhouse gases than its other fossil fuel cousins.

### Investigating Changes in Natural Gas Prices
The investigation in this section is to determine the month that natural gas price witnessed the most frequent changes and to explore reasons for these changes. In order to do this, a new values for every row is created with one of three possible values: “Price-Drop”, “Price-Rise, or “No Change”” This can be done by creating two derived columns:

1. Based on the difference between the present gas price and the previous day’s price (Price-Difference)
2. Assigned ‘tags’ in the newly created Price-Difference column and call it ‘Price-Change’

These two new columns are known as a derived columns because they have been created using data from one or more existing columns.

Add the Price-Difference Column by assigning values to the column name. Show the first 3 rows.

```python
df_gas_prices['Price-Difference'] = 'Difference'
df_gas_prices[:3]
```

Replace the value ‘Difference’ with the value derived from computing the differences of consecutive rows for the HHNGS-Price ($/mmbtu) column. Use the shift method.

```python
df_gas_prices['Price-Difference'] = df_gas_prices['Gas Prices ($/mmbtu)'] - df_gas_prices['Gas Prices ($/mmbtu)'].shift(-1)
print (df_gas_prices.head())
```

Front fill the last value in the new column created which is NAN.

```python
df_gas_prices['Price-Difference'].fillna(0, inplace=True)
print (df_gas_prices.tail())
```

To create a column “Price-Change” with status of the the price differences assigned:
Change data type of ‘Price-Difference’column from string to numeric

```python
df_gas_prices[['Price-Difference']] = df_gas_prices[['Price-Difference']].apply(pd.to_numeric)
```

Define a function (f(row) that operates on the rows of the dataframe using conditional statements and pass this to apply and set axis=1.

```python
def f(row):
    if row['Price-Difference'] == 0.00:
        val =  "No-Change"
    elif row['Price-Difference'] > 0.00:
        val = "Price-Rise"
    else:
        val = "Price-Drop"
    return val#
df_gas_prices['Price-Change'] = df_gas_prices.apply(f, axis=1)
print(df_gas_prices.head())
```

```python
df_gas_prices['Price-Change'] = df_gas_prices.apply(f, axis=1)
print (df_gas_prices.head())
```

We now want to investigate the months of the year when there is more price changes than others. This will enable us to determine whether there are special features of such months that determine price changes . 

To compare price changes by months, we need to group the monthly records together. We will use the .groupby() function which allows us to group records into buckets by categorical values. Since we already have created a column in our data for the Price-Change, and also another column to indicate the month of the price change, we can simply pass those arguments into the groupby() function. ##### 2005-2006
Disruptions caused by severe weather, operating mishaps, or planned maintenance can also cause short-term tightness in natural gas supply. In the summer of 2005, hurricanes along the U.S. Gulf Coast (Katrina/Rita, and the 2005 Loss of Offshore Supply) caused more than 800 billion cubic feet (Bcf) of natural gas production to be shut in between August 2005 and June 2006. This is equivalent to about 5 percent of U.S. production over that period and about 22 percent of yearly natural gas production in the Federal Gulf of Mexico. As a result of these disruptions, natural gas spot prices at times exceeded $15 per million Btu (MMBtu) in many spot market locations and fluctuated significantly over the subsequent months, reflecting the uncertainty over supplies.

### Histogram
We can visually represent the distribution of gas price changes using a histogram. Histograms allow us to bucket the values into bins, or fixed value ranges, and count how many values fall in that bin.
numpy .arange()
We are interested in analyzing size of gas price changes and we want to put these sizes into bins that represent every 10 cents/MMbtu increase. We can use the numpy method .arange() to create a list of numbers that define those bins. The bins of 10 cents/MMbtu intervals will range from $10/MMbtu price drop (-10) to 10 cents/MMbtu price increase (10). The first bin will hold a count of gas price change between $10MMbtu and $9.90/MMbtu drop in price, then $9.90/MMbtu and $9.80/MMbtu, and so on.

```python
bin_values = np.arange(start=-1.0, stop=1.1, step=0.010)
print(bin_values)
```

# Fitting Linear Regression using statsmodels
Statsmodels is a Python library for a lot of basic and inferential statistics. It also provides basic regression functions using an R-like syntax, so it's commonly used by statisticians. 

```python
# Import regression modules
# ols - stands for Ordinary least squares, we'll use this
import statsmodels.api as sm
from statsmodels.formula.api import ols
```

```python
# statsmodels works nicely with pandas dataframes
# The thing inside the "quotes" is called a formula
g = ols('GPRICE ~ GPROD',df_data).fit()
print (g.summary())
```

#### Interpreting coefficients

There is a ton of information in this output. But we'll concentrate on the coefficient table (middle table). We can interpret the `GPROD` coefficient (-2.814e-06) by first noticing that the p-value (under `P>|t|`) is basically zero. We can interpret the coefficient as, if we compare the natural gas production at two periods of time, one where the gas throughput is say 200000 MMcf and the other period is the same except that they all have a throughput of 3000000 MMscf. For these two periods the difference in Natural Gas Spot price is about marginal (-$2.814e-06), which is basically Zero. The confidence interval gives us a range of plausible values for this difference, about (-$3.81e-06 and  -$1.82e-06), definitely not a remarkable change. 

#### Linear Regression with Gas Export Price
Let us use statsmodels to carry out linear regression on gas export prices

```python
# statsmodels works nicely with pandas dataframes
# The thing inside the "quotes" is called a formula
m = ols('GPRICE ~ GEXPP',df_data).fit()
print (m.summary())
```

#### Interpreting coefficients

Just as described before, looking at the coefficient table, we can interpret the `GEXPP` coefficient (1.0733) by first noticing that the p-value (under `P>|t|`) is zero. This means that if we compare the export prices of natural gas at two periods of time, one where the export price is say $3 per MMscf and the other period is the same except its price is $4 per MMScf. For these two periods the difference in Natural Gas Spot price is about $1.1. The confidence interval gives us a range of plausible values for this difference, about ($1.0 and  $1.1).

##### Summary
Here is a summary of what we did: We 've loaded in the data, split it into a training and testing sets, fitted a regression model to the training data, made predictions based on this data and tested the predictions on the test data. 

However, the train/test split does have its dangers: Such as — What if the split we make isn’t random? What if one subset of our data has only data from a certain year, or the data is from certain features. This will result in overfitting, even though we’re trying to avoid it! This is where cross validation comes in.

#### Cross Calidation
In this section, we apply the train/test split to more subsets. Meaning, we split our data into k subsets, and train on k-1 one of those subset. What we do is to hold the last subset for test. We’re able to do it for each of the subsets.

We will use K-Folds Cross Validation and the Leave One Out Cross Validation (LOOCV)

##### K-Folds Cross Validation
In K-Folds Cross Validation we split our data into k different subsets (or folds). We use k-1 subsets to train our data and leave the last subset (or the last fold) as test data. We then average the model against each of the folds and then finalize our model. After that we test it against the test set.



```python
from sklearn.model_selection import KFold # import KFold
kf = KFold(n_splits=2) # Define the split - into 2 folds 
kf.get_n_splits(X) # returns the number of splitting iterations in the cross-validator
print(kf)
```

And let’s see the result — the folds:

```python
for train_index, test_index in kf.split(X):
    print('TRAIN:', train_index, 'TEST:', test_index)
  #  X_train, X_test = X[train_index], X[test_index]
   # Y_train, Y_test = Y[train_index], Y[test_index]
```

As we can see, the function splits the original data into different subsets of the data. 

#### Leave One Out Cross Validation (LOOCV)
In this type of cross validation, the number of folds (subsets) equals to the number of observations we have in the dataset. We then average ALL of these folds and build our model with the average. We then test the model against the last fold. Because we would get a big number of training sets (equals to the number of samples), this method is very computationally expensive but we are using it because our dataset is small datasets. 

```python
from sklearn.model_selection import LeaveOneOut 
loo = LeaveOneOut()
loo.get_n_splits(X)
```

```python
for train_index, test_index in loo.split(X):
    print("TRAIN:", train_index, "TEST:", test_index)
```

The more folds we have, we will be reducing the error due the bias but increasing the error due to variance; the computational price would go up too, obviously — the more folds we have, the longer it would take to compute it and we would need more memory. With a lower number of folds, we’re reducing the error due to variance, but the error due to bias would be bigger. It would also computationally cheaper. Therefore, in big datasets, k=3 is usually advised. In smaller datasets it’s best to use LOOCV.

We will use the cross_val_predict function to return the predicted values for each data point when it’s in the testing slice.

```python
# Necessary imports: 
from sklearn.cross_validation import cross_val_score, cross_val_predict
from sklearn import metrics
```

Recall that we ’ve created the train/test split for our dataset and fitted a model. Let’s see what the score of our model is after cross validation:

```python
# Perform 6-fold cross validation
scores = cross_val_score(lm.fit(X_train, Y_train), X, Y, cv=6)
print ('Cross-validated scores:', scores)
```

As we can see, the last fold improved the score of the original model — from 0.964755825429 to 1. 
Now, let’s plot the new predictions, after performing cross validation:

```python
# Make cross validated predictions
predictions = cross_val_predict(lm.fit(X_train, Y_train), X, Y, cv=6)
plt.scatter(Y, predictions)
```

We can see our scatter plot is very different from the original plot (its now a near perfect correlation). It is six times as many points as the original plot because we used cv=6.

Finally, let’s check the R² score of the model (R² is a “number that indicates the proportion of the variance in the dependent variable that is predictable from the independent variable(s)”. Basically, how accurate is our model):

```python
accuracy = metrics.r2_score(Y, predictions)
print ('Cross-Predicted Accuracy:', accuracy)
```
