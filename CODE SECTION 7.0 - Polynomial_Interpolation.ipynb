```python
import pandas as pd
# Read the data set
df=pd.read_excel("data/DataSet_GasPrice_ Outlier_Removed.xlsx")
df.head()
```

```python
# create new dataframe of independent variables
dfcoal = df['AveCoalPrice']
```

```python
# create numpy arrays of the dependent and independent variables with n_samples, n_features = 104, 4
#X = dfvar.values
X = dfcoal.values
Y= df.GasPrice.values
```

```python
def f(X):
    """ function to approximate by polynomial interpolation"""
    return X * np.sin(X)


# generate points used to plot
x_plot = np.linspace(0, 10, 104)

# generate points and keep a subset of them
X= np.linspace(0, 10, 104)
rng = np.random.RandomState(0)
rng.shuffle(X)
X = np.sort(X[:104])
y = f(X)
```

```python
# create matrix versions of these arrays
X = X[:, np.newaxis]
X_plot = x_plot[:, np.newaxis]
```

```python
X.shape
```

```python
colors = ['teal', 'yellowgreen', 'gold']
lw = 2
plt.plot(X_plot, Y, color='cornflowerblue', linewidth=lw,
         label="ground truth")
plt.scatter(X, Y, color='navy', s=30, marker='o', label="training points")

for count, degree in enumerate([5, 7, 9]):
    model = make_pipeline(PolynomialFeatures(degree), Ridge())
    model.fit(X, Y)
    y_plot = model.predict(X_plot)
    plt.plot(X_plot, y_plot, color=colors[count], linewidth=lw,
             label="degree %d" % degree)

plt.legend(loc='lower left')

plt.show()
```
